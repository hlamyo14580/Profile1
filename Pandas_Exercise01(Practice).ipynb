```python
#* How many people have the word Chief in their job title? (This is pretty tricky) **
def chief_string(title):
    if 'chief' in title.lower():
        return True
    else:
        return False
```

```python
sal[sal['JobTitle']=='%chief%']
```

```python
chief_string('CHIEF MANAGER-METROPOLITAN TRANSIT AUTHORITY')
```

```python
def chief(x):
    if 'chief' in x.lower().split():
        return True
    else:
        return False
```

```python
sal['JobTitle'].apply(chief).sum()
```

```python
df=pd.DataFrame(sal['JobTitle'].str.contains('Chief'))
counts=df.value_counts()
print(counts)
```
